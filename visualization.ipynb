```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
def Plot_res(SA,TC,Gen,Save=False):
    cm = plt.cm.get_cmap('copper')
    fig, ax = plt.subplots( figsize = (4,3))
    plt.rc('font',family='Arial',weight='normal')
    plt.rcParams.update({'font.size':14})
    font1 = {'family' : 'Arial',
    'weight' : 'normal',
    'size'   : 16,
    }
    font2 = {'family' : 'Arial',
    'weight' : 'normal',
    'size'   : 16,
    }
    ax_plot=plt.scatter(SA,
                      TC,
                     c=Gen,cmap=cm, marker='o',s=3, alpha=0.8)
    title='SA'
    title1='TC (W m$^{-1}$K$^{-1}$)'
    plt.xlabel(title,font1 )
    plt.ylabel(title1,font1 )
    plt.xlim((2,4.5))
    plt.ylim((0, 1.10))
    my_x_ticks = np.arange(2,4.6,0.5)
    my_y_ticks = np.arange(0, 1.02,0.2)
    plt.xticks(my_x_ticks,size=14)
    plt.yticks(my_y_ticks,size=14)
    plt.tick_params(width=1.5)
    bwith = 1.5 #边框宽度设置为2
    TK = fig.gca()#获取边框
    TK.spines['bottom'].set_linewidth(bwith)#图框下边
    TK.spines['left'].set_linewidth(bwith)#图框左边
    TK.spines['top'].set_linewidth(bwith)#图框上边
    TK.spines['right'].set_linewidth(bwith)#图框右边 
    fig.colorbar(ax_plot,ax=ax,label='Generations')
    if Save:
        plt.savefig(r'Candiadates.png', dpi=300,transparent=True,bbox_inches = 'tight')
    plt.tight_layout()
    plt.show()
```

```python
def Plot_HV(gen,HV,Save=False):
    fig,ax = plt.subplots(figsize=(3.5,3))
    plt.rc('font',family='Arial',weight='normal')
    font1 = {'family' : 'Arial',
    'weight' : 'normal',
    'size'   : 16,
    }
    font2 = {'family' : 'Arial',
    'weight' : 'normal',
    'size'   : 16,
    }       

    ax_plot = plt.plot(gen,HV,color='r',linestyle='-',linewidth=1.5,zorder=2)
    title='Generations'
    title1='Hypervolumes'
    plt.xlabel(title,font1 )
    plt.ylabel(title1,font1 )
    plt.xlim((0,50))
    plt.ylim((0, 8.0))
    my_x_ticks = np.arange(0,51,10)
    my_y_ticks = np.arange(0,8.02,2)
    plt.xticks(my_x_ticks,size=14)
    plt.yticks(my_y_ticks,size=14)
    plt.tick_params(width=1.5)
    bwith = 1.5
    TK = plt.gca()
    TK.spines['bottom'].set_linewidth(bwith)
    TK.spines['left'].set_linewidth(bwith)
    TK.spines['top'].set_linewidth(bwith)
    TK.spines['right'].set_linewidth(bwith)
    if Save:
        plt.savefig(r'HVs.png', dpi=300,transparent=True,bbox_inches = 'tight')
    plt.tight_layout()
```

```python
# Result for MOGA
df_GA=pd.read_csv(r'./results/MOEA_candidates.csv')
df_GA_HV=pd.read_csv(r'./results/MOEA_HV.csv')
SA_ga=df_GA['SA']*-1
TC_ga=df_GA['TC']
Gen_ga=df_GA['Gen']
HV_ga=np.array(df_GA_HV)
print('==================Results for MOEA==================')
Plot_res(SA_ga,TC_ga,Gen_ga)
Plot_HV(HV_ga[:,0],HV_ga[:,1])
```

```python
# Result for MObO
df_BO=pd.read_csv(r'./results/MOBO_candidates.csv')
df_BO_HV=pd.read_csv(r'./results/MOBO_HV.csv')
SA_bo=df_BO['SA']*-1
TC_bo=df_BO['TC']
Gen_bo=df_BO['Gen']
HV_bo=np.array(df_BO_HV)
print('==================Results for MOBO==================')
Plot_res(SA_bo,TC_bo,Gen_bo)
Plot_HV(HV_bo[:,0],HV_bo[:,1])
```
